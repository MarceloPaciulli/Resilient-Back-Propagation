```python
# -*- coding: utf-8 -*-

import numpy as np
from matplotlib import pylab
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import latexify
import warnings


#suppress warnings 'return 1/(1 + np.exp(-x))'
warnings.filterwarnings('ignore')


class Identity(object):                   
    def __call__(self, x):               
        return x

    def deriv(self, x):
        return 1


class Sigmoid(object):
    def __call__(self, x):
        return 1/(1 + np.exp(-x))

    def deriv(self, x):
        n = self(x)
        return n * (1 - n)


FUNCS = {
    'Identity': Identity,
    'Sigmoide': Sigmoid,
} 


#ANN: Artificial Neural Network
class ANN(object):   
    def __init__(self, inputs, hidden, outputs, func_h=Sigmoid(), func_o=Identity()):
        self.inputs = inputs
        self.hidden = hidden
        self.outputs = outputs
        self.func_h = func_h
        self.func_o = func_o
        self.restart() 
       
    def restart(self, delta=0.5):
        delta2 = delta * 2
        self.W_h = np.random.random((self.inputs, self.hidden)) * delta2 - delta 
        self.b_h = np.random.random(self.hidden) * delta2 - delta                  
        self.W_o = np.random.random((self.hidden, self.outputs)) * delta2 - delta  
        self.b_o = np.random.random(self.outputs) * delta2 - delta                  
       
    @staticmethod         
    def load(file_name):             
        data = np.load(file_name)

        ann = ANN(data['inputs'], data['ocultas'], data['outputs'], 
                  func_h=FUNCS[str(data['func_h'])](),                  
                  func_o=FUNCS[str(data['func_o'])]())                  
        ann.W_h = data['W_h']
        ann.b_h = data['b_h']
        ann.W_o = data['W_o']
        ann.b_o = data['b_o']

        return ann

    def save(self, file_name):
        np.savez(file_name,
              inputs=self.inputs, outputs=self.outputs, hidden=self.hidden,
              func_h=self.func_h.__class__.__name__, func_o=self.func_o.__class__.__name__,
              W_h=self.W_h, b_h=self.b_h, W_o=self.W_o, b_o=self.b_o)

    def _evaluate_inside(self, X):
        z_h = np.dot(X, self.W_h) + self.b_h 
        y_h = self.func_h(z_h)               
        z_o = np.dot(y_h, self.W_o) + self.b_o
        y_o = self.func_o(z_o)              
        return (z_h, y_h, z_o, y_o)         
                                            
    def evaluate(self, X):
        return self._evaluate_inside(X)[-1] 


    def train_rprop(self, X, T, min_error=0,
                       delta_0=0.1, delta_min=1e-6, delta_max=50.0,
                       eta_minus=0.5, eta_plus=1.2,
                       max_iter=500, callback=None):

        prev_grad = [np.zeros_like(self.W_h), 
                     np.zeros_like(self.b_h),  
                     np.zeros_like(self.W_o),  
                     np.zeros_like(self.b_o),  
        ]

        deltas = [np.empty_like(self.W_h),
                  np.empty_like(self.b_h), 
                  np.empty_like(self.W_o),
                  np.empty_like(self.b_o), 
        ]
        for d in deltas:
            d.fill(delta_0)

        weights = [self.W_h, self.b_h, self.W_o, self.b_o]

        for n in range(max_iter):
            z_h, y_h, z_o, y_o = self._evaluate_inside(X)
            err = T - y_o
            E = 0.5 * (err ** 2).sum()

            if callback is not None:
                callback(self, X, T, y=y_o, n=n+1, E=E)

            if E <= min_error:
                break

            delta_o = err * self.func_o.deriv(z_o)
            delta_h = self.func_h.deriv(z_h) * np.dot(delta_o, self.W_o.T)
            grad = [(delta_h[:, np.newaxis, :] * X[:, :, np.newaxis]).sum(axis=0),    # W_h
                    delta_h.sum(axis=0),                                              # b_h
                    (delta_o[:, np.newaxis, :] * y_h[:, :, np.newaxis]).sum(axis=0),  # W_o
                    delta_o.sum(axis=0),                                              # b_o
            ]

            for pg, g, d, w in zip(prev_grad, grad, deltas, weights):
                cond_ne_0 = g != 0
                w[cond_ne_0] += np.sign(g[cond_ne_0]) * d[cond_ne_0]
                cond = pg * g
                cond_gt_0 = cond > 0
                cond_lw_0 = cond < 0
                d[cond_gt_0] = np.minimum(d[cond_gt_0] * eta_plus, delta_max)
                d[cond_lw_0] = np.maximum(d[cond_lw_0] * eta_minus, delta_min)
                g[cond_lw_0] = 0

            prev_grad = grad

        return E, n + 1

@latexify.function
def f(x):
    return x*np.sin(x)
    
@latexify.function
def g(x):
    return (1/x)*np.sin(x)

@latexify.function
def h(x):
    return np.exp(-x**2)

@latexify.function
def threshold(x):
    return 1/(1 + np.exp(-x))
    
ann = None

def function_approx_example(f):
    global ann
    print ('Example of how to use an ANN as an approximation of an arbitrary function:  \n')
    X = np.r_[-4 * np.pi: 4 * np.pi: 300j].reshape((-1, 1))
    T = f(X) #example function: f(x) = x*sin(x)          
    X2 = np.r_[-4 * np.pi: 4 * np.pi: 1000j].reshape((-1, 1))  
    X4 = np.r_[ 0 * np.pi: 2 * np.pi: 1000j].reshape((-1, 1))

    def progress(ann, X, T, y=None, n=-1, E=None):
        print("Iteration: {0} - Error: {1:.4f} - Y.min(): {2} - Y.max(): {3}".format(n, E, y.min(), y.max()))
        if n > 0 and n % 100 == 0:    
            print('#'*94)
            output_plot(X, T, X, y)

    def output_plot(X, T, X2, y):         
        pylab.plot(X, T, 'b.', X2, y, 'r')    
        pylab.draw()
        
    hidden = 500
    inputs = X.shape[1] 
    outputs = T.shape[1]  
  
    ann = ANN(inputs, hidden, outputs, func_h=Sigmoid(), func_o=Identity())
    y = ann.evaluate(X)
    E, n = ann.train_rprop(X, T, min_error=0.002, max_iter=1500, callback=progress)
    print("Network trained in {0} iterations, with an error of {1:6.3f}".format(n, E))
    y4 = ann.evaluate(X4)
    output_plot(X, T, X4, y4)
    plt.title("ANN responses and original function {0}".format(r'$'+str(f)+'$'))
    ann_response = mlines.Line2D([], [], color='red',
                          markersize=15, label='ANN responses')
    function_plot = mlines.Line2D([], [], color='blue',
                          markersize=15, marker='.', label=r'$'+str(f)+'$')
    plt.legend(handles=[ann_response, function_plot])
    plt.show()
    
#functions: f, g, h, threshold
function_approx_example(f)

    

```
